```python
# install libraries
import pandas as pd
from pycaret.regression import *
```

```python
# set the maximum number of column to display:
pd.set_option('display.max_columns', 100)

#  datload our dataset ->> in this case "housing_price_prediction.csv"
housing = pd.read_csv("housing_price_prediction.csv")
housing.head()
```

```python
# Explore the dataset
housing.shape
```

```python
housing.dtypes.value_counts()
```

```python
# lest check missing value 
missing_values = housing.isnull().sum()
missing_columns = missing_values[missing_values>0]

for col, count in missing_columns.items():
    print(f"Column {col}) has {count} missing values")
```

```python
# initialize the environment and prepare data for modeling
caret_df = setup(data = housing, target = "SalePrice", session_id= 123)

# session_id: you can put any random number it wont mean anything. In this case i put 123.

# if you want to explictily define the categorical variable while initialize module you can do it.
# But in this case pycared will automatically do it.
# categorical_features = ["MSZoning", "Street", "...", "..."]

# if you think you want to avoid some features to ignore in our model you can explictily declare it. 
# ignore_features=[‘total_production’]
```

```python
# The two_words
# now time to compare different regression model with our dataset. 
best = compare_models()
```

```python
# Analyze the model
evaluate_model(best)
```

```python
# plot the model in confusion matrix
predict_model(best)
```

```python
tuned_freg = tune_model(best)
```
